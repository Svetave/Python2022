#Task 1

Prosite database for Isocitrate lyase: K-[KR]-C-G-H-[LMQR]

<br>

**Isocitrate lyase (FASTA ID) contains its active site XXXXXX starting from Nth position**


```python
import re
```

```python
# import txt file with Isocytrate in fasta format
import requests
res=requests.get("https://raw.githubusercontent.com/Svetave/PIK3CA/main/Isocytrate.txt")
file=res.text
print(file)
#https://raw.githubusercontent.com/Svetave/PIK3CA/main/Isocytrate.txt
```

```python
def site(file):
  s=file.split("\r")# clean sequence
  clean_text=[]
  for i in s:
      new=i.strip("\n")
      clean_text.append(new.split(" "))
  clean_text
  sequence=[]
  for i in clean_text[1:]:
    for j in i:
      sequence.append(j)
  clean_sequence=("").join(sequence)
  header=clean_text[0][0] # header for FASTA ID
  header=("").join(re.findall(r"\w+.+\w", header))
  pattern=r"K[K|R]CGH[L|M|Q|R]"
  res = re.search(pattern, clean_sequence)# find position of pattern
  position=res.start()+1 
  site="".join(re.findall(pattern, clean_sequence))# find pattern
  print(f"Isocitrate lyase ({header}) contains its active site {site} starting from {position}th position")# print result

site(file)
```

```python
res2=requests.get("https://raw.githubusercontent.com/Svetave/Python2022/main/Employees_insert.txt")
employ=res2.text
```

```python
def clean(employ):
  final=[]
  clean_list=[]
  zub=[]
  fo=[]
  clean_text=[]
  sequence=[]
  s=employ.split("\r")# clean sequence
  for i in s:
      new=i.strip("\n")
      clean_text.append(new.split(" "))
  for i in clean_text[3:]:
    for j in i:
      sequence.append(j)
  clean_sequence2=("").join(sequence)
  clean_sequence2=clean_sequence2.split(r"INSERTINTOemployeesVALUES")

  for t in clean_sequence2:
    final.append(t.split(","))
  for line in final:
    clean_list.append(line[1:3])# keep only three first values
  for k in clean_list:
    k=(",").join(k)
    line = re.sub('\W+',' ', k) # remove all non-word character
    zub.append(line)

  paty=re.compile(r"(N[e|a]+\w+) (\w+)") #search for pattern
  for i in zub:
    a=paty.findall(i)
    if a != []:
      fo.append(" ".join(a[0]))
  return fo

print(clean(employ))
```

```python
#3.1
def street(location):
  emp1=[]
  sequence3=[]
  clean_sequence3=[]
  streets=location.split("\r\nINSERT INTO locations VALUES \r\n        ")# clean sequence
  streets=streets[1:]
  pattern_street=re.compile(r"([\w\w+.\w+,+\s]+\s+\w+[\w\s\-]*)")
  for g in streets:
      strnames=re.findall(pattern_street,g)[0]
      strnames = re.sub('\s','',strnames) # remove all spaces
      emp1.append(strnames)
  return emp1
rty=street(location)
rty
```

```python
#3.2
def street_loc(location):
  emp=[]
  sequence4=[]
  clean_sequence4=[]
  streets4=location.split("\r\nINSERT INTO locations VALUES \r\n        ")# clean sequence
  streets4=streets4[1:]
  pattern_street4=re.compile(r"([\d\w\w+.\w+,+\s]+\s+\w+[\w\s\-]*)")
  for g in streets4:
      strnames4=pattern_street4.findall(g)[0]
      emp.append(strnames4)
  st_adr=[]
  for gg in emp:
    strnames5=re.findall(r"([\d\w\s]*[St|Strt]$)",gg)
    if strnames5 != []:
      line4 = re.sub('St$','Street',strnames5[0])
      st_adr.append(line4)
  return(st_adr)
street_loc(location)
```

```python
number1="909-90-99"
number2="(892)909-90-99"
number3="33344-33"
number4="(892)sssssss"
number5="(892)909-9099"

def check_number(number):
  reg4=re.compile(r"^\d{3}\-\d{2}\-\d{2}$")
  reg5=re.compile(r"^\(\d{3}\)\d{3}\-\d{2}\-\d{2}$")
  if reg4.search(number) or reg5.search(number):
    print("It's a number!")
  else:
    print("It's not a number!")

check_number(number1)
check_number(number2)
check_number(number3)
check_number(number4)
check_number(number5)
```
