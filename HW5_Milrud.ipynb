## HW 5 Functional programming

1. `itertools.product()`

https://www.hackerrank.com/challenges/itertools-product/problem?isFullScreen=true

```python
from itertools import product

A=list(map(int,input().strip().split()))
B=list(map(int,input().strip().split()))
pr=product(A,B)
[print(x,end=" ") for x in pr ]
```

2. `itertools.permutations(iterable[, r])`

https://www.hackerrank.com/challenges/itertools-permutations/problem?isFullScreen=true

```python
from itertools import permutations

per_input=input().strip().split()

clean_list=sorted(list(permutations(per_input[0],int(per_input[1]))))
for i in clean_list:
    print(("").join(i))
```

3. `itertools.combinations(iterable, r)`

https://www.hackerrank.com/challenges/itertools-combinations/problem?isFullScreen=true

```python
from itertools import combinations

com_input=input().strip().split()

for i in range(1,int(com_input[1])+1):
    for j in combinations(sorted(com_input[0]),i):
        print("".join(j))
```

4. `itertools.combinations_with_replacement(iterable, r)`

https://www.hackerrank.com/challenges/itertools-combinations-with-replacement/problem?isFullScreen=true

```python
from itertools import combinations_with_replacement

com_input=input().strip().split()
clean_list=sorted(com_input[0])
for i in combinations_with_replacement(clean_list,int(com_input[1])):
    print(("").join(i))
```

5. Compress the string

https://www.hackerrank.com/challenges/compress-the-string/problem?isFullScreen=true

```python
from itertools import groupby

S=list(map(int,input()))

for i,j in groupby(S):
    print(tuple([len(list(j)), i]) ,end = " ")
```

6. Iterables and iterators

https://www.hackerrank.com/challenges/iterables-and-iterators/problem?isFullScreen=true

```python
from itertools import combinations

N=int(input())
N_list=input().split()
K=int(input())
a='a'
comb_list=list(combinations(N_list,K))
count=0
for i in comb_list:
    if a in i:
        count+=1
print(count/len(comb_list))
```

7. Maximize

https://www.hackerrank.com/challenges/maximize-it/problem?isFullScreen=true

```python
from itertools import product

K,M = map(int,input().split())

N = (list(map(int, input().split()))[1:] for i in range(K))

result = 0
for i in product(*N):
    r = 0
    for n in i:
        r += n**2       
    if result < r % M:
        result = r % M
print(result)


```
